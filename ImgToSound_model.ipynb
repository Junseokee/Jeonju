```python
import librosa
import librosa.display
import torch
import torchaudio
from glob import glob
import IPython.display as ipd
import time

label_dic = {
'0' : '다듬잇돌', '1' : '교회 종', '2' : '괘종시계', 
'3' : '먹 갈기' , '4' : '떡메', '5': '맷돌', '6':'물레방아', '7' : '석유풍로',
'8' : '숫돌', '9' : '싸리비', '10' : '가위질', '11': '엿 자르기', '12': '작두', '13': '절구', 
'14' : '타자기', '15' : '키', '16' :'풍로', '17': '확독'}

file_list=glob("./runs/detect/**/labels/")
result_folder = file_list[-1]

label_txt_list=glob(result_folder+"*.txt", recursive=True)

for i in label_txt_list:
    audio_path = "./sound/"
    with open(i,'r') as f:
        line=f.readline()
        label = label_dic.get(line.split(' ')[0])
    
    audio, sr = librosa.load(audio_path+label+'.wav')
print('이 사진은 '+ label+'소리입니다.')
ipd.Audio(audio, rate = sr, autoplay=True)
```
